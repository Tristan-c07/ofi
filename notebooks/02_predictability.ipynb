```python
import pandas as pd
import os

# 定义 tick 数据路径
df = r"data/processed/ticks/159915.XSHE/2020-01-02/part.parquet"

df = pd.read_parquet(df)

df['ts'] = pd.to_datetime(df['ts'])

print(df.head())
```

```python
df['minute'] = df['ts'].dt.floor('min')

close = (df.groupby('minute').last()['a1_p'] + df.groupby('minute').last()['b1_p'] )/ 2

ret = close.shift(-1) / close - 1
print(ret.head())
print(close.head())
```
